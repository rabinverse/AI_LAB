```python
import pandas as pd 
import numpy as np
import os
import matplotlib.pylab as plt
```

```python
artha_banija_path = '../../dataset/Nepali_news/ArthaBanijya/ArthaBanijya/'
bichar_path = '../../dataset/Nepali_news/Bichar/Bichar'
desh_path = '../../dataset/Nepali_news/Desh/Desh'
khelkud_path = '../../dataset/Nepali_news/Khelkud/Khelkud'
manoranjan_path = '../../dataset/Nepali_news/Manoranjan/Manoranjan'
prabas_path = '../../dataset/Nepali_news/Prabas/Prabas'
sahitya_path = '../../dataset/Nepali_news/Sahitya/Sahitya'
suchana_prabidhi_path = '../../dataset/Nepali_news/SuchanaPrabidhi/SuchanaPrabidhi'
swasthya_path = '../../dataset/Nepali_news/Swasthya/Swasthya'
viswa_path = '../../dataset/Nepali_news/Viswa/Viswa'
```

```python
def read_datafiles(path):
    total_text = []
    for filename in os.listdir(path):
        if filename.endswith('.txt'):
            with open(os.path.join(path,filename),'r') as file:
#                 
                tokens = path.split('/')
                label = tokens[-1]      
                text = file.read()
                total_text.append((text,label))
    return total_text
```

```python
artha_banija = read_datafiles(artha_banija_path)
bichar = read_datafiles(bichar_path)
desh = read_datafiles(desh_path)
khelkud = read_datafiles(khelkud_path)
manoranjan = read_datafiles(manoranjan_path)
prabas = read_datafiles(prabas_path)
sahitya = read_datafiles(sahitya_path)
suchana_prabidhi = read_datafiles(suchana_prabidhi_path)
swasthya = read_datafiles(swasthya_path)
viswa = read_datafiles(viswa_path)
```

```python
X = df["text"]
Y = df["label"]
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
from sklearn.pipeline import Pipeline, make_pipeline
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    ConfusionMatrixDisplay,
)
```

```python
tfidf = TfidfVectorizer()
```

```python
nb_model = make_pipeline(TfidfVectorizer(stop_words="english"), MultinomialNB())
```

```python
type(xtest)
```

```python
nb_model.fit(xtrain,ytrain)
```

```python
y_pred=nb_model.predict(xtest)
```

```python
print("Accuracy",accuracy_score(ytest,y_pred))
print(classification_report(ytest,y_pred))
```
